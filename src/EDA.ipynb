```python
import pandas as pd
import numpy as np

dta = pd.read_csv("Final_Datasets/not_preprocessed.csv")
```

```python
dta["prelec_correct"] = np.where(dta["prelec_prediction"] == dta["actual"], 1,0)
dta["major_correct"] = np.where(dta["majority_prediction"] == dta["actual"], 1,0)
balance = dta.groupby(["actual"]).size()[0]/210
major_acc = sum(dta["major_correct"])/210
prelec_acc = sum(dta["prelec_correct"])/210
```

```python
d = {'Balance': [balance], 'Major Accuracy': [major_acc], 'Prelec Accuracy': [prelec_acc], 'Logistic': [0.75], 'SVM': [0.77], 'Random Forrest': [0.81], 'Nearest Neighbor': [0.76]}
df = pd.DataFrame(data=d)
df = df.T
df.columns = ["Accuracy Score"]
df["Model Types"] = df.index
```

```python
import seaborn as sns
sns.set(style="whitegrid")
ax = sns.barplot(df["Model Types"], df["Accuracy Score"], data=df)
ax.set_title('Accuracy Scores for Different Models')
ax.set_xticklabels(rotation=90,labels=df["Model Types"])
```
